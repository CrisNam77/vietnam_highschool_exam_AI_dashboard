## 1. Bức tranh nguồn dữ liệu thô

Bộ dữ liệu trải qua bốn năm nhưng lại chứa đựng hai sự thay đổi vô cùng lớn: việc đổi mới chương trình giáo dục phổ thông (áp dụng từ năm 2025) và việc sáp nhập đơn vị hành chính cấp tỉnh (từ 63 tỉnh về 34 tỉnh, áp dụng trong bộ dữ liệu từ năm 2026). Do đó, bảy tệp dữ liệu thô thu thập được không hề đồng nhất và có thể chia làm ba nhóm cấu trúc chính:

1. **Nhóm CSV 2022-2024:** Định dạng CSV, sử dụng tên cột dạng `snake_case` (ví dụ: `toan`, `ngu_van`), và chứa điểm của chương trình giáo dục 2006.
2. **Nhóm Excel 2025:** Định dạng XLSX, phân tách thành hai chương trình (2006 và 2018), sử dụng tên cột là tiếng Việt có dấu (ví dụ: `Toán`, `Lí`).
3. **Nhóm CSV 2026:** Định dạng CSV, chương trình 2018, sử dụng tên cột tiếng Việt có dấu, bổ sung thêm cột `Tỉnh` có sẵn và dùng hệ mã tỉnh mới sau sáp nhập.

## 4. Quy trình tiền xử lý

Dưới đây là tóm tắt quy trình chạy ngầm bên trong script `clean_data.py`. Pipeline được thiết kế để chuẩn hóa như sau:

| Bước | Việc thực hiện | Nguyên tắc giữ gìn |
|------|----------------|---------------------|
| **1. Chuẩn hóa cột** | Đổi toàn bộ tên cột thô về bộ tên chung (canonical). | Thêm các cột thiếu với giá trị rỗng hoặc mặc định. |
| **2. Làm sạch SBD** | Ép kiểu chuỗi, xóa khoảng trắng, đệm đủ 8 ký tự. | Bỏ các SBD chứa chữ cái hoặc không đủ chiều dài. |
| **3. Ánh xạ địa lý** | Quy đổi toàn bộ dữ liệu về **hệ thống 34 tỉnh mới**. <br>- <= 2025: Map từ 2 số đầu của SBD.<br>- 2026: Lấy và chuẩn hóa từ cột Tỉnh. | Lọc bỏ các tỉnh lạ không nằm trong danh mục chuẩn. |
| **4. Xử lý Điểm** | Ép kiểu số thực (float) cho 13 môn thi. Loại bỏ các giá trị nằm ngoài khoảng hợp lệ. | Giữ nguyên điểm liệt 0 và điểm tuyệt đối 10. Không thay trống bằng 0. Không cắt điểm ngoại lai. |
| **5. Biến dẫn xuất** | Tạo thêm các cột phụ trợ: `so_mon`, `ban` (cho CT2006), `diem_khoi_a00`... | Rỗng nếu thiếu dù chỉ một môn trong khối. Không tự ý xếp ban cho CT2018. |
| **6. Lọc dữ liệu** | Bỏ các thí sinh có số môn thi bằng 0. | Tôn trọng các thí sinh đăng ký nhưng chỉ thi 1 môn. |

## 5. Kết quả sau xử lý

Dữ liệu đầu ra là một tệp duy nhất, được lưu dưới định dạng bảng ngang (WIDE), với tên cột thống nhất, số báo danh tròn trịa, tên tỉnh chuẩn xác, và sẵn sàng cho công tác phân tích chuyên sâu.


```python
# Đọc tập dữ liệu sau khi đã được làm sạch
final_df = pd.read_csv(
    "../data/processed/final_data.csv",
    dtype={"sbd": str, "ma_tinh": str, "ma_ngoai_ngu": str,
           "chuong_trinh": str, "ban": str},
)
display(f"Kích thước bộ dữ liệu: {final_df.shape}")
display("Mẫu dữ liệu sau xử lý:")
display(final_df.head())
```

```python
# Phân bố số dòng theo Năm và Chương trình
dist_df = final_df.groupby(['nam', 'chuong_trinh']).size().reset_index(name='Số lượng dòng')
display("Phân bố dữ liệu thành phẩm theo năm và chương trình:")
display(dist_df)
```

```python
# Thống kê chất lượng sau xử lý
quality_stats = pd.DataFrame({
    "Hạng mục": [
        "Tổng dòng thô đầu vào",
        "Số dòng vắng thi toàn bộ (đã bị loại)",
        "Số dòng báo danh sai định dạng (loại)",
        "Số dòng mã tỉnh lạ ngoài danh mục (loại)",
        "Số ô điểm nằm ngoài khoảng [0, 10] (set rỗng)",
        "Tổng dòng WIDE dataset thành phẩm"
    ],
    "Số lượng": ["5.364.307", "4.637", "0", "0", "0", "5.359.670"]
})
display("Bảng báo cáo chất lượng dữ liệu cuối cùng:")
display(quality_stats)
```

**Nhận xét:** 
Đối chiếu với dữ liệu nguyên bản từ Mục 2, sau khi xử lý:
- Các tên cột lộn xộn đã được đồng nhất về bộ canonical.
- Các cột địa lý (`ten_tinh`, `vung_mien`, `vung_3`) đã được chuẩn hóa về bộ 34 tỉnh duy nhất, sẵn sàng phục vụ biểu đồ vùng miền xuyên suốt 5 năm.
- Điểm thi đã mang định dạng số và loại bỏ những dòng vắng thi hoàn toàn.
- Chất lượng đầu vào xuất sắc (không phát hiện SBD hay điểm lỗi định dạng).

## 8. Lưu ý khi sử dụng dữ liệu

1. **Mã tỉnh là Nơi dự thi:** Dữ liệu phản ánh địa điểm thí sinh thi chứ không đại diện cho nơi cư trú của thí sinh. Do đó, các chênh lệch địa phương cần được diễn đạt thận trọng.
2. **Hệ thống 34 tỉnh bản lề:** Vì dữ liệu các năm cũ (2022-2025) đã được map ép sang hệ thống 34 tỉnh hiện hành, hình thù vùng miền ở thời điểm quá khứ sẽ trở nên "thô" hơn so với thực tế và một vài địa phương (như Kon Tum, Bình Thuận) đã bị chuyển dịch khu vực kinh tế. Nhóm phân tích khuyên chỉ groupby theo `ten_tinh` thay vì `ma_tinh`.
3. **Mặt bằng đánh giá biến động mạnh:** Năm 2025 và 2026 là giai đoạn chuyển tiếp chương trình mới. Sự thay đổi cấu trúc đề thi dẫn tới việc các chỉ số trung bình và xếp hạng không có chung một hệ quy chiếu thước đo so với ba năm 2022-2024.
4. **Không suy luận nhân quả:** Dữ liệu cung cấp duy nhất hai chiều thông tin: Điểm thi và Vị trí địa lý. Việc rút ra bất cứ nguyên nhân nào liên quan tới chất lượng sống, cơ sở hạ tầng, hay năng lực quản lý đều mang tính suy đoán, không được chứng minh qua bộ dữ liệu này.
